```python
import requests
import pandas as pd

# API URL (example endpoint)
url = "https://keyvalue.hamropatro.com/kv/get/major-election-2082-states-1-result"

# headers (important for some APIs)
headers = {
    "User-Agent": "Mozilla/5.0",
    "Accept": "application/json"
}

# request data
response = requests.get(url, headers=headers)

# convert to json
data = response.json()

print("API Response Received")

# extract actual value
value = data.get("value")

# sometimes value itself is JSON string
import json
parsed_data = json.loads(value)

# store rows here
rows = []

for area in parsed_data:
    
    area_name = area.get("name")

    for candidate in area.get("candidates", []):
        
        row = {
            "Area": area_name,
            "Candidate": candidate.get("name"),
            "Party": candidate.get("party"),
            "Votes": candidate.get("votes")
        }

        rows.append(row)

# convert to pandas dataframe
df = pd.DataFrame(rows)

# save csv
df.to_csv("nepal_election_results.csv", index=False)

print("\nData Saved to CSV Successfully")
print(df.head())
```

```python
import requests
import pandas as pd
import json

# API URL
url = "https://keyvalue.hamropatro.com/kv/get/major-election-2082-states-1-result"

headers = {
    "User-Agent": "Mozilla/5.0",
    "Accept": "application/json"
}

response = requests.get(url, headers=headers)
data = response.json()

value = data.get("list")[0].get("value")  # extract the JSON string

if value is None:
    print("⚠️ No data available")
else:
    parsed_data = json.loads(value)  # now we have a Python dict

    rows = []
    for state in parsed_data.get("stateResults", []):
        area_name = state.get("areaName")
        district_name = state.get("districtName")
        state_name = state.get("stateName")
        total_counted_votes = state.get("totalCountedVotes")
        total_cast_votes = state.get("totalCastVotes")
        result_status = state.get("electionResultStatus")

        for candidate in state.get("candidateResults", []):
            rows.append({
                "State": state_name,
                "District": district_name,
                "Area": area_name,
                "Candidate Name": candidate.get("name"),
                "Candidate English Name": candidate.get("englishName"),
                "Party": candidate.get("partyName"),
                "Votes": candidate.get("votes"),
                "Winner": candidate.get("winner"),
                "Featured": candidate.get("featured"),
                "Result Final": candidate.get("resultFinal"),
                "Image URL": candidate.get("imageUrl"),
                "Slug": candidate.get("slug"),
                "Total Counted Votes": total_counted_votes,
                "Total Cast Votes": total_cast_votes,
                "Result Status": result_status
            })

    # Convert to DataFrame
    df_koshi = pd.DataFrame(rows)

    # Save to CSV
    df_koshi.to_csv("nepal_election_results.csv", index=False)
    print("✅ CSV saved with", len(df), "rows")
```
